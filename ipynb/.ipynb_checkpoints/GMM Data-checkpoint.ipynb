## KDD cup 2004

- Link to data: http://localhost:8888/notebooks/ipynb/GMM%20Data.ipynb#KDD-cup-2004
- Description: predicting whether proteins are homologous to the native sequence or decoys
- Dimensions: 145,751 observations in a 74-dimensional space
- Source: page 21 of [Qiong's paper](https://arxiv.org/pdf/2010.10412.pdf); also page 11 of [Lucic et al. (2017)](https://arxiv.org/pdf/1703.08110.pdf)

Pros:
- used in various papers for GMM fitting

Cons:
- hard to fit, high-dimensional (papers it is used in are for coresets)

```python
kdd_dat = pd.read_table('dat/data_kddcup04/bio_train.dat',header=None)
```

```python
kdd_dat
```

## Magic 04

- Link to data: https://archive.ics.uci.edu/dataset/159/magic+gamma+telescope
- Description: registration of high energy gamma particles in a ground-based atmospheric Cherenkov gamma telescope using the imaging technique
- Dimensions: 19,020 observations in a 10-dimensional space
- Source: page 21 of [Qiong's paper](https://arxiv.org/pdf/2010.10412.pdf)

Pros:
- Not as high-dimensional. Observations large but can be subsampled

Cons:
- Data were simulated, not real

```python
magic_dat = pd.read_table('dat/magic+gamma+telescope/magic04.data',sep=None,header=None)
```

```python
magic_dat
```

## Waveform

- Link to data: https://hastie.su.domains/ElemStatLearn/datasets/waveform.train
- Description: Simulated waveform data
- Dimensions: 300 observations in a 21-dimensional space
- Source: page 451 of [Elements of Statistical Learning](https://hastie.su.domains/Papers/ESLII.pdf) by Hastie et al. Also featured in page 4 of [Liu et al. (2010)](https://ojs.aaai.org/index.php/AAAI/article/view/7659)

Pros:
- Well studied
- Small number of observations

Cons:
- Data were simulated, not real

```python
waveform_dat=pd.read_table('https://hastie.su.domains/ElemStatLearn/datasets/waveform.train')
```

```python
waveform_dat
```

- Dimensions: 990 observations in a 10-dimensional space
